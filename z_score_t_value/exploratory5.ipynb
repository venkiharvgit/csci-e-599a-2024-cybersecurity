```python
z_block164 = []
for p in range(100000):
  z_block164.append(z_vals164)
z_block164 = np.array(z_block164)
```

```python
def get_DF_values2e(aDF):
    tuples_list = []
    columns = aDF.columns
    selected_columns = columns[4:-2]  # Excluding first 4 columns, columns 22 and 23, and last 2 columns
    selected_columns = selected_columns.drop(columns[[21, 22]])  # Excluding columns 22 and 23
    for index, row in aDF.iterrows():
        values = row[selected_columns]
        tuples_list.append(np.array(values))
    tuples_list = np.array(tuples_list)
    return tuples_list

def normalizer(a_value,amin,amax):
    return (a_value - amin) / (amax - amin)

def calculate_t_values(z_scores_first_data, row_values_second_data):

    denom = np.sqrt(1/len(row_values_second_data))

    passer =  np.array([(row_values_second_data[p] - z_scores_first_data[p])/denom for p in range(row_values_second_data.shape[0])])
    a_min = min(passer)
    a_max = max(passer)
    return np.array([normalizer(item,a_min,a_max) for item in passer])
```

```python
gepa = zepa + '/NF-ToN-IoT-v2_CH_88.parquet'
enDF1 = pd.read_parquet(gepa)
enDF1.head(10)
```

```python
grp1 = get_DF_values2e(enDF1)
t_values_88e = []
for row in grp1:
  en_pass = calculate_t_values(z_vals164,row)
  t_values_88e.append(en_pass)
```
